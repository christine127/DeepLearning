```python
# from keras.boston_housing import load_data
from sklearn.datasets import load_boston
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import load_iris
warnings.filterwarnings(action= 'ignore')

np.random.seed(121)
tf.random.set_seed(121)

dataset = load_boston()
X= dataset.data
y = dataset.target

df =pd.DataFrame(data = dataset.data, columns = dataset.feature_names)
df["price"]= dataset.target
print(df.shape)
print(df.info())
```

```python
#-------------------------------------------model 
model =Sequential()
model.add(Dense(units=128, input_dim=13, activation ='relu'))
model.add(Dense(units=64, activation ='relu'))
model.add(Dense(units=16, activation ='relu'))
model.add(Dense(units=1)) #h(x) = wx + b
model.compile(loss='mean_squared_error',optimizer ='adam')

#------------------------------------------train,test set
# y_log = np.log1p(y)

X_train,X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=160)
model.fit(X_train,y_train, epochs= 100)
pred= model.predict(X_val)
for i in range(5):
    print(y_val[i], pred[i])
    
```
